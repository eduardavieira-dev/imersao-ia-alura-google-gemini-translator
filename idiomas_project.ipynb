```python
import google.generativeai as genai
from google.colab import userdata
#Cole sua API KEY abaixo dessa maneira: api_key = 'SUA_API_KEY'
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)

# Configurações do modelo
generation_config = {
  "temperature": 0.5,
  "top_p": 0.95,
  "top_k": 35,
  "max_output_tokens": 1000,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
]


model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

# Função para traduzir com Gemini e fornecer exemplos e explicação
def traduzir_gemini(texto, idioma_destino):
  prompt_traducao = f"Por favor, reescreva o seguinte texto em {idioma_destino}: '{texto}'"
  prompt_exemplos = f"Por favor, forneça 3 exemplos de como a tradução de '{texto}' para {idioma_destino} pode ser usada em frases."
  prompt_explicacao = f"Explique o significado da tradução de '{texto}' para {idioma_destino}."

  convo = model.start_chat()
  convo.send_message(prompt_traducao)
  traducao = convo.last.text

  convo.send_message(prompt_exemplos)
  exemplos = convo.last.text

  convo.send_message(prompt_explicacao)
  explicacao = convo.last.text

  return traducao, exemplos, explicacao

while True:
  # Obtém o texto do usuário
  texto = input("Digite o texto que deseja traduzir (ou 'fim' para encerrar): ")

  # Verifica se o usuário digitou "fim"
  if texto.strip().lower() == "fim":
    break

  # Obtém o idioma de destino do usuário
  print("Digite o número do idioma para o qual deseja traduzir e ver exemplos na língua:")
  print("1 - Português")
  print("2 - Francês")
  print("3 - Inglês")
  print("4 - Espanhol")
  opcao_idioma = input("Digite o número da opção: ")

  if opcao_idioma == "1":
    idioma_destino = "Português"
  elif opcao_idioma == "2":
    idioma_destino = "Francês"
  elif opcao_idioma == "3":
    idioma_destino = "Inglês"
  elif opcao_idioma == "4":
    idioma_destino = "Espanhol"
  else:
    print("Opção inválida. Tente novamente.")
    continue # Volta para o início do loop

  # Realiza a tradução, obtém exemplos e explicação
  traducao, exemplos, explicacao = traduzir_gemini(texto, idioma_destino)

  # Imprime os resultados
  print(f"\nTexto original: {texto}")
  print(f"Idioma de destino: {idioma_destino}\n")
  print(f"Tradução: {traducao}")
  print(f"Exemplos: {exemplos}")
  print(f"Explicação: {explicacao}")

print("Programa encerrado.")
```
